В качестве набора данных я выбрала датасет про размер устриц. В таблице представленны данные о массе и объёме каждой устрицы (в унциях). На графике же хорошо видно, как размер устрицы зависит от этих двух параметров

```python
import matplotlib.pyplot as pt
import pandas as pd
import numpy as np
import scipy as sp
import math

trainSet=pd.read_csv('input/Oysters.csv')
print(pd.DataFrame(trainSet).head(7))
```

Подключение всех библиотек, загрузка данных в переменные, формирование фреймов

```python
df = pd.DataFrame(trainSet)
X=df["Weight"].tolist()
Y=df["Volume"].tolist()
i=df["№"].tolist()
```

Выделение фреймов на сеты X и Y

```python
pt.title('Oysters size ')
pt.scatter(X,Y,c='black',marker='*')
pt.xlabel('Weight, oince')
pt.ylabel('Volume, fluid oince')
pt.show()
```

```python
bar_width = 0.5
pt.title('Oysters size bar diagram')
pt.bar(np.arange(len(i)), X, bar_width, color='grey', label='Weight')
pt.bar(np.arange(len(i)) + bar_width, Y, bar_width, alpha=0.8, color='k', label='Volume')
pt.ylabel("oince")
pt.legend()
pt.show()
```

```python
th1_1, th0_1 = sp.polyfit(X, Y, 1)
th2_2, th1_2, th0_2 = np.polyfit(X, Y, 2)
th3_3, th2_3, th1_3, th0_3 = np.polyfit(X, Y, 3)
th4_4, th3_4, th2_4, th1_4, th0_4 = np.polyfit(X, Y, 4)
x = [1, len(Y)]
m = list(range(len(Y)))
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 1))(m), label='1 degree polynomial')
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 2))(m), label='2 degree polynomial')
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 3))(m), label='3 degree polynomial')
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 4))(m), label='4 degree polynomial')
pt.legend()
pt.show()
```

Из графика следует, что зависимость размера устрицы - линейная. Возьмём задачу посложнее. Например, построим график выживаемости крыс в зависимости от выброшенной дозы радиции. За предоставленные дата-сеты отдельное спасибо https://vincentarelbundock.github.io/Rdatasets/datasets.html

```python
trainSet=pd.read_csv('input/survival.csv')
df = pd.DataFrame(trainSet)
X=df["dose"].tolist()
Y=df["surv"].tolist()
i=df["№"].tolist()
```

```python
pt.title('Survival of Rats after Radiation Doses')
pt.scatter(X,Y,c='black',marker='*')
pt.xlabel('dose')
pt.ylabel('survived, %')
pt.show()
```

```python
th1_1, th0_1 = sp.polyfit(X, Y, 1)
th2_2, th1_2, th0_2 = np.polyfit(X, Y, 2)
th3_3, th2_3, th1_3, th0_3 = np.polyfit(X, Y, 3)
th4_4, th3_4, th2_4, th1_4, th0_4 = np.polyfit(X, Y, 4)
x = [1, len(Y)]
m = list(range(len(Y)))
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 1))(m), label='1 degree polynomial')
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 2))(m), label='2 degree polynomial')
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 3))(m), label='3 degree polynomial')
pt.plot(m, sp.poly1d(np.polyfit(X, Y, 4))(m), label='4 degree polynomial')
pt.legend()
pt.show()
```

Вывод: Крысы предсказуемо мрут при повышении дозы радиации, и чем выше степень полинома, тем утопичнее его прогнозы.
